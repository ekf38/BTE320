```python
#Income Statement

#input
sales = float(input("Enter Sales ($):"))
tax_rate = float(input("Enter Tax Rate (%):"))
#process
COGS = sales * 0.4
Profit = sales - COGS
taxes = Profit * (tax_rate / 100)
net_income = Profit - taxes
#output
print("Sales: ${:.2f}".format(sales))
print("Cost Of Goods Sold: ${:.2f}".format(COGS))
print("Gross Profit: ${:.2f}".format(Profit))
print("Taxes: ${:.2f}".format(taxes))
print("Net Income: ${:.2f}".format(net_income))
```

```python
#Simple Income Statement Calculation

'''
demand (in units per year)
reorder cost($/order)
holding costs($/unit/year
EOQ = sqrt=(2*D*S/H)
Looking to minimize EOQ\
'''

demand = int(input("Enter Demand (units/year):"))
reorder_cost = float(input("Enter Reorder Cost ($/order):"))
holding_cost = float(input("Enter Holding Cost ($/unit/year):"))

EOQ = int(((2 * demand * reorder_cost) / holding_cost)**0.5)

print("EOQ: {} units".format(EOQ))
```

```python
#Rock Paper Scissor AI
import random

def play_game():
    choices = ["rock", "paper", "scissors"]

    while True:
        # Ask user if they want to play
        play = input("Do you want to play Rock, Paper, Scissors? (yes/no): ").strip().lower()
        if play != "yes":
            print("Thanks for playing! Goodbye!")
            break

        # Get user choice
        user_choice = input("Choose rock, paper, or scissors: ").strip().lower()
        if user_choice not in choices:
            print("Invalid choice, please try again.")
            continue

        # Computer randomly chooses
        computer_choice = random.choice(choices)
        print(f"The computer chose: {computer_choice}")

        # Decide winner
        if user_choice == computer_choice:
            print("It's a tie!")
        elif (user_choice == "rock" and computer_choice == "scissors") or \
             (user_choice == "paper" and computer_choice == "rock") or \
             (user_choice == "scissors" and computer_choice == "paper"):
            print("You win!")
        else:
            print("Computer wins!")

        print()  # blank line for readability


# Run the game
play_game()
```

```python
#Assignment II Using Exception Handling

def getValues(n: int) -> list:
  items = []
  for i in range(n):
    item = float(input("Enter item: "))
    items.append(item)

  return items

def calcAverage(alist: list) -> float:
  try:
    return sum(alist[2:]) / len(alist[2:])
  except ZeroDivisionError:
    print('The list is empty')
    return (float('nan'))

def main():
  n = int(input("Enter number of items: "))
  if n < 0:
    raise ValueError("Number of items cannot be negative")

  L = getValues(n)
  print(f"Average: {calcAverage(L)}")

main()
```

```python
#Assignment Loan & NPV Mini-Toolkit

def roi(cost: float, revenue: float) -> float:
  assert cost > 0, "Cost must be greater than 0"
  assert isinstance(cost, (int, float)), "Cost must be a number"
  assert isinstance(revenue, (int, float)), "Revenue must be a number"

  return float(revenue - cost) / cost

if __name__ == "__main__":
  c = float(input("Enter cost: "))
  r = float(input("Enter revenue: "))
  ret = roi(c, r)
  print(ret)
```

```python
#Monthly Loan Payment

def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float)), "Principal must be numeric."
  assert isinstance(annual_rate, (int, float)), "Annual rate must be numeric."
  assert isinstance(months, (int, float)), "Months must be numeric."

  if principal < 0:
    raise ValueError("Principal must be a positive number.")
  if months < 0:
    raise ValueError("Months must be a positive number.")
  if annual_rate < 0:
    raise ValueError("Annual rate cannot be negative.")

  monthly_rate = annual_rate / 12

  if monthly_rate < 0:
    pmt = monthly_rate * principal / (1 - (1 + monthly_rate) ** (-months))
  else:
    pmt = principal / months

  return pmt

def main():
    try:
        principal = float(input("Enter the loan principal: "))
        annual_rate = float(input("Enter the annual interest rate (as a decimal): "))
        months = int(input("Enter the loan term in months: "))
        payment = loan_payment(principal, annual_rate, months)
        print(f"The monthly loan payment is: ${payment:.2f}")
    except (ValueError, AssertionError) as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

main()
```

```python
#Greeting Class

class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def __str__(self):
    return self.first + ' ' + self.last

  def display(self, reverse=True):
    if reverse:
      print(f'{self.last}, {self.first}')
    else:
      print(self.first)

name = Greeting('John', 'Doe')
print(name)
name.display()
name.display(reverse=False)
```

```python
#Student Class

class Student:
  def __init__(self, name, number):
    self.__name = name
    self.__number = number
    self.__courses = []

  def getName(self):
    return self.__name

  def getCourses(self):
    return self.__courses

  def setName(self, newname):
    self.__name = newname

  def __len__(self):
    return len(self.__courses)

  def enroll(self, course):
    if course in self.__courses:
      print('You have already enrolled in', course)
    else:
      self.__courses.append(course)

n = Student('Bob', 12345)
print(n.getCourses())
print(n.getName())
n.setName('John')
n.enroll('ABC123')
n.enroll('ABC123')
print(n.getCourses())
len(n)
```

```python
#class point and rectangle

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    return((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(1,0)
    self.p3 = Point(1,1)
    self.p4 = Point(0,1)
    self.width = self.p1.distance(self.p2)
    self.height = self.p3.distance(self.p2)

  def area(self):
    return self.height * self.width

square = Rectangle()
square.area()
```
